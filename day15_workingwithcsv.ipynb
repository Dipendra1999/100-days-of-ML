#convertor function on values of the specific column
we can change the name of values in column
eg:RoyalchalengerBanglore to RCB

using the
# convertors={"column_name": _convertor_function}

```python
def rename(name):
  if name=="Royal Challengers Bangalore":
    return "RCB"
  else:
    return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
df=pd.read_csv("/content/ipl-matches-2008-2020/IPL Matches 2008-2020.csv",parse_dates=["date"],converters={"team1":rename})
```
